```python
from scipy import optimize
import math
import numpy as np
from scipy import linalg

%matplotlib inline
from matplotlib import pylab as plt
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
w = linalg.solve(a, b)

def x1(x):
    return w[0] + w[1] * x

x = np.arange(1, 15, 1)
y = list(map(lambda x: f(x), x))
yPol = list(map(lambda x: x1(x), x))

plt.plot(x, y, '-', x, yPol, '-')
plt.show()
```

```python
a = np.array([[1, 1, 1], [1, 8, 8 ** 2], [1, 15, 15 ** 2]])
b = np.array([f(1), f(8), f(15)])
w = linalg.solve(a, b)

def x2(x):
    return w[0] + w[1] * x + w[2] * (x ** 2)

x = np.arange(1, 15, 1)
y = list(map(lambda x: f(x), x))
yPol = list(map(lambda x: x2(x), x))

plt.plot(x, y, '-', x, yPol, '-')
plt.show()
```

```python
a = np.array([[1, 1, 1, 1], [1, 4, 4 ** 2, 4 ** 3], [1, 10, 10 ** 2, 10 ** 3], [1, 15, 15 ** 2, 15 ** 3]])
b = np.array([f(1), f(4), f(10), f(15)])
w = linalg.solve(a, b)

def x3(x):
    return w[0] + w[1] * x + w[2] * (x ** 2) + w[3] * (x ** 3)

x = np.arange(1, 15, 1)
y = list(map(lambda x: f(x), x))
yPol = list(map(lambda x: x3(x), x))

plt.plot(x, y, '-', x, yPol, '-')
plt.show()
```
